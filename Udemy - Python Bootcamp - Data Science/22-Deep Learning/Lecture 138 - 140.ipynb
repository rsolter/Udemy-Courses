### Modeling

```python
X = df.drop('price',axis=1)
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
X_train = scalar.fit_transform(X_train) # scale and fit at the same 
X_test = scalar.fit_transform(X_test) 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(19,activation='relu')) 
model.add(Dense(19,activation='relu')) 
model.add(Dense(19,activation='relu')) 
model.add(Dense(19,activation='relu')) 

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,
        validation_data=(X_test,y_test),
        batch_size=128, # usually fed in powers of 2
        epochs=400) 
# can track performance vs test data by use of validation =
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot() # we want to see similar losses in validation and test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions) # off by 114k
```

```python
df['price'].describe()
```

```python
# Mean absolute error is around 20% not horrible, not great
```

```python
explained_variance_score(y_test,predictions) # 1 is max
```

```python
plt.figure(figsize=(11,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r') # adding a perfect fit line in red

# Notice that we appear to be punished by outliers. 
```

```python
## Retraining on bottom 99% of houses
# re creating bottom 99% from engineered df

bottom_99_percent = df.sort_values('price',ascending=False).iloc[216:]
```

```python
X = bottom_99_percent.drop('price',axis=1)
y = bottom_99_percent['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scalar = MinMaxScaler()
```

```python
X_train = scalar.fit_transform(X_train) # scale and fit at the same 
X_test = scalar.fit_transform(X_test) 
```

```python

model = Sequential()

model.add(Dense(19,activation='relu')) 
model.add(Dense(19,activation='relu')) 
model.add(Dense(19,activation='relu')) 
model.add(Dense(19,activation='relu')) 

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
model.fit(x=X_train,y=y_train,
        validation_data=(X_test,y_test),
        batch_size=128, # usually fed in powers of 2
        epochs=400) 

losses = pd.DataFrame(model.history.history)
losses.plot() # we want to see similar losses in validation and test
```

```python
predictions = model.predict(X_test)
np.sqrt(mean_squared_error(y_test,predictions))
mean_absolute_error(y_test,predictions) # off by 111k
bottom_99_percent['price'].describe() # off by less than 20%
```

```python
explained_variance_score(y_test,predictions) # 1 is max
```

```python
plt.figure(figsize=(11,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r') # adding a perfect fit line in red

# Notice that we appear to be punished by outliers. 
```
