## Broadcasting Operations

```python
# Similar to apply as it was used on Series
nba = pd.read_csv("nba.csv")
nba.head()
```

```python
nba["Age"].add(5) # accounts for nulls

# Alternative approach that pandas is smart enough to handle
nba["Age"] + 5
```

```python
# Subtracting 5 million from each salary
nba["Salary"].sub(-5000000)
nba["Salary"] - 5000000
```

```python
# Converting weights into kGs

nba["Weight"].mul(0.453)
nba["Weight"] * 0.453
```

```python
# Adding to a new column
nba["Weight in Kilograms"] = nba["Weight"] * 0.453 
nba
```

```python
nba["Salary in millions"] = nba["Salary"].div(1000000) 
```

## Drop Rows with Null Values

```python
nba = pd.read_csv("nba.csv")
nba
```

```python
nba.dropna() # drops any value with a NA
```

```python
nba.dropna(how="all", inplace=True) #only removes rows with NAs in all values
```

```python
nba.tail()
```

```python
# Remove columns with any na values -- drops Salary and College
nba.dropna(axis=1)
```

```python
# Only dropping observations where there are nulls in "Salary" column
nba.dropna(subset=["Salary"]) # Drops John Holland
```

## Fill in Null Values with the .fillna() Method

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.fillna(value=0) # will replace all NAs within the datafram with 0 (not ideal)
```

```python
# Better to apply it by column/series
nba["Salary"].fillna(0,inplace=True)
```

```python
nba
```

```python
nba["College"].fillna("No college",inplace=True)
```

```python
nba
```

## The .astype() Method

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"].fillna(0,inplace=True)
nba["College"].fillna("No college",inplace=True)
nba
```

```python
nba.dtypes
```

```python
nba.info # Should be giving size of object in (kB) 
```

```python
nba["Salary"].astype("int") # converting float to int, there's not an Inplace=True argument, however
```

```python
nba["Salary"] = nba["Salary"].astype("int") 
```

```python
nba.head()
```

```python
# Category data type
# Like a 'factor' in R, uses much less space than a string
```

```python
nba["Position"].nunique() # Ideal candidate, not too many unique values
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
nba["Tean"] = nba["Team"].astype("category")
```

```python
nba.info
```

## Rank Values with the .rank() Method

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba
```

```python
nba["Salary"].rank()
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
```

```python
nba
```

```python
nba.sort_values(by="Salary",ascending=False)
```
